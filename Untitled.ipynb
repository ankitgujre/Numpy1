```python
size = 1000
list1 = range(size)
list2 = range(size)
start = time.time()
result = [x+y for x,y in zip(list1,list2)]
end = time.time()
print("Python list time", end - start)

arr1 = np.arange(size)
arr2 = np.arange(size)
start = time.time()
result = arr1+arr2
end = time.time()
print("Python arr time", end - start)
```
